```python
import sympy as sp
import random

x = sp.Symbol('x')

def constdiff():
    """differentiate a constant"""

    # avoid 0 because it's too trivial
    const_val = random.choice([i for i in range(-50,51) if i != 0])

    diff = sp.diff(const_val, x)
    diff_latex = sp.latex(diff)

    steps = []
    steps.append("<problem>")
    steps.append(f"solve: $\\frac{{d}}{{dx}}({const_val})$")
    steps.append("</problem>")

    steps.append("<solution>")
    steps.append(f"step 1: identify the function: $f(x) = {const_val}$")
    steps.append("step 2: apply constant rule: $\\frac{d}{dx}(c) = 0$")
    steps.append(f"step 3: differentiate: $\\frac{{d}}{{dx}}({const_val}) = {diff_latex}$")
    steps.append(f"final answer: $\\frac{{d}}{{dx}}({const_val}) = {diff_latex}$")
    steps.append("</solution>")

    return '\n'.join(steps)
```

```python
import sympy as sp
import random

x = sp.Symbol('x')

def onevariable():
    """differentiate a linear term ax"""

    const = random.choice([i for i in range(-50, 51) if i != 0])
    expr = const * x
    expr_latex = sp.latex(expr)
    diff = sp.diff(expr, x)
    diff_latex = sp.latex(diff)

    steps = []

    steps.append("<problem>")
    steps.append(f"solve: $\\frac{{d}}{{dx}}({expr_latex})$")
    steps.append("</problem>")

    steps.append("<solution>")
    steps.append(f"step 1: identify the function: $f(x) = {expr_latex}$")
    steps.append("step 2: apply identity rule: $\\frac{d}{dx}(x)=1$")
    steps.append(f"step 3: differentiate: $\\frac{{d}}{{dx}}({expr_latex}) = {const}$")
    steps.append(f"final answer: $\\frac{{d}}{{dx}}({expr_latex}) = {diff_latex}$")
    steps.append("</solution>")

    return '\n'.join(steps)
print(onevariable())
```

```python
def func_derivative(func):
    derivatives = {
        r'\sin{(x)}': r'\cos{(x)}',
        r'-\sin{(x)}': r'-\cos{(x)}',
        r'\cos{(x)}': r'-\sin{(x)}',
        r'-\cos{(x)}': r'\sin{(x)}',
        r'\tan{(x)}': r'\sec^{2}{(x)}',
        r'-\tan{(x)}': r'-\sec^{2}{(x)}',
        r'\csc{(x)}': r'-\csc{(x)}\cot{(x)}',
        r'-\csc{(x)}': r'\csc{(x)}\cot{(x)}',
        r'\sec{(x)}': r'\sec{(x)}\tan{(x)}',
        r'-\sec{(x)}': r'-\sec{(x)}\tan{(x)}',
        r'\cot{(x)}': r'-\csc^{2}{(x)}',
        r'-\cot{(x)}': r'\csc^{2}{(x)}'
    }
    return derivatives.get(func)
```

```python
def sinx():
    funcs = [
        r'\sin{(x)}', r'\cos{(x)}', r'\tan{(x)}', r'\csc{(x)}', r'\sec{(x)}', r'\cot{(x)}',
        r'-\sin{(x)}', r'-\cos{(x)}', r'-\tan{(x)}', r'-\csc{(x)}', r'-\sec{(x)}', r'-\cot{(x)}'
    ]
    function = random.choice(funcs)
    diff = func_derivative(function)

    steps = []
    steps.append('<problem>')
    steps.append(f'solve: $\\frac{{d}}{{dx}}({function})$')
    steps.append('</problem>')
    steps.append('<solution>')
    steps.append(f'step 1: identify the function: $f(x) = {function}$')
    steps.append(f'step 2: apply derivative rule: $\\frac{{d}}{{dx}}({function}) = {diff}$')
    steps.append(f'step 3: final answer: ${diff}$')
    steps.append('</solution>')

    return '\n'.join(steps)

print(sinx())
```
